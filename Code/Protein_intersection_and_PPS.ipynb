```python
from Bio import SeqIO
from Bio import Align
import pandas as pd
from tqdm import tqdm
```

```python
protein_dataset=pd.read_csv('drive/MyDrive/protein_inner_join.csv')
aligner = Align.PairwiseAligner()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
protein_dataset.head(-5)
```

```python
Dict = {}
```

```python
for i in tqdm(range(len(protein_dataset))):
    for j in range(len(protein_dataset)):
        if(i==j):
            break
        Score = aligner.score(protein_dataset['sequence'][i],protein_dataset['sequence'][j])
        Dict[(protein_dataset['id'][i],protein_dataset['id'][j])]=Score/max(len(protein_dataset['sequence'][i]),len(protein_dataset['sequence'][j]))
```

```python
len(Dict)
```

```python
protien_sim=[]
for i in Dict:
  row=[]
  if (Dict[i]>0.5):
    row.append(i[0])
    row.append(i[1])
    row.append(Dict[i])
    protien_sim.append(row)
```

```python
protien_sim[0:20]
```

```python
protien_similarity_df_split = pd.DataFrame(protien_sim, columns=['compound0', 'compound1', 'similarity'])
```

```python
protien_similarity_df_split.head(-5)
```

```python
protien_similarity_df_split.to_csv('drive/MyDrive/protien_similarity_threshold_df.csv', index=False)
```
